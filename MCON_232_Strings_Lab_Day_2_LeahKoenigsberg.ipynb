```python
students = []

# write file
with open("students.txt", "w") as fptr:
    fptr.write("Eli\n20\nProf Margolis\n")
    fptr.write("Alana\n18\nProf Spiegel-Cohen\n")
    fptr.write("Rina\n21\nProf Novick\n")

# read + group
#define variables
one_student = ""
count = 0
#read in the data and write on one line
with open("students.txt", "r") as fptr:
  for student_data in fptr:
    #you will read and write it with this variable
    one_student += student_data
    #for each student there's name, age and prof  so keep count to 3
    count += 1

    #so when it hits 3, append to list
    if count == 3:
      students.append(one_student)
      one_student = ""
      count = 0

#after file closes, print the list created
for student in students:
  print(f'{student.replace('\n', ' ')}')
```

```python
#read in my csv file - parsing the data
#read books.csv without using csv library
#will pass in str- name of file and then return a tuple within a list...
def read_books_no_csv(filename : str = "books.csv") -> list[tuple[str, str, int, str, str]] :
  fin = open(filename, "r")
  lines = fin.readlines()
  fin.close()

  books = []
  for line in lines[1:]:
    line = line.strip()

    if line == "":
      continue

    #parts is a list
    parts = line.split(",")

    #here we are relying on where the data is:
    title = parts[0].strip()
    author = parts[1].strip()
    year = int(parts[2].strip())
    country = parts[3].strip()
    genre = parts[4].strip()

    books.append((title, author, year, country, genre))
  return books

csv_reader = read_books_no_csv()

for line in csv_reader:
  print(line)
```

```python
def count_lines(filename):
  with open(filename, "r") as read_file:
    count  = 0

    for line in read_file:
      lines = line.lower()
      lines = line.strip()

      #pnct = ",./;'\":?[]{}()!-_"
      #for i in pnct:
      new = lines.replace(".", " ")
      new_line = new.split(" ")

      for i in new_line:
        if i.startswith("pale") or i == "pallor":
          count+=1

    return count



count_lines("pg345_cleaned.txt")
```

```python
def clean_to_words(line):
   # Replace common punctuation with spaces, then split
   punctuation = ".,;:!?\"'()[]{}-_—"
   line = line.lower()
   for ch in punctuation:
       line = line.replace(ch, " ")
   return line.split()


def is_pale_word(word):
   # matches pale, pales, paled, paleness
   if word.startswith("pale"):
       return True
   # matches pallor exactly
   if word == "pallor":
       return True
   return False


def count_pale_lines(filename):
   fin = open(filename, "r", encoding="utf-8")
   count = 0


   for line in fin:
       words = clean_to_words(line)
       found = False
       for w in words:
           if is_pale_word(w):
               found = True
               break
       if found:
           count += 1


   fin.close()
   return count
count_pale_lines("pg345_cleaned.txt")
```
