```python
def clean_sentence(s: str) -> str:
  s = s.strip()
  s= s.lower()
  new = s.replace('!', '.')
  new = new.replace('?', '.')
  lst = new.split()
  final = " ".join(lst)
  return final

print(clean_sentence("  Hello!!!   HOW are you?? "))


def count_vowels(s: str) -> int:
  vowels = "aeiou"
  count = 0
  for i in s.lower():
    if i in vowels:
      count+=1
  return count
print(count_vowels("Education"))


def reverse_words(sentence: str) -> str:
  s = sentence.split()
  reversed = s[::-1]
  reversed = " ".join(reversed)
  return reversed
print(reverse_words("Python is fun"))


def first_unique_char(s: str) -> str:
  new = ""
  for i in range(len(s)):
    if s[i] not in new:
      new += s[i]
    else:
      new = new.replace(s[i], "")
  for j in new:
    return j
  return None
print(first_unique_char("swiss"))


def replace_digits(s: str) -> str:
  digits = {
    "0": "zero",
    "1": "one",
    "2": "two",
    "3": "three"
  }
  final=""
  for i in s:
    if i in digits:
      final+= digits[i]
    else:
      final+= i
  return final

print(replace_digits("I have 2 apples and 1 orange"))


def word_lengths(sentence: str) -> list:
  sentence = sentence.split()
  new_list = []
  num = 0
  for i in range(len(sentence)):
    for j in range(len(sentence[i])):
      num += 1
    new_list.append(num)
    num = 0
  return new_list
print(word_lengths("Python is awesome"))


def find_letter_positions(word: str, letter: str) -> list:
  count = 0
  lst = []
  for i in range(len(word)):
    if letter == word[i]:
      lst.append(i)
  return lst
print(find_letter_positions("banana", "a"))


def remove_duplicates(s: str) -> str:
  new = ""
  for i in range(len(s)):
    if s[i] not in new:
      new += s[i]
  return new
print(remove_duplicates("banana"))


def is_palindrome(s: str) -> bool:
  s= s.lower()
  s = s.split()
  new = "".join(s)
  final = new[::-1]
  if final == new:
    return True
  return False
print(is_palindrome("Race car"))


def format_names(names: list) -> str:
  cleaned = []
  for i in range(len(names)):
    name = names[i].strip()
    name = name.replace("  ", " ")
    name = name.title()
    cleaned.append(name)
  final = ", ".join(cleaned)
  return final


print(format_names([" mary  smith ", "JOHN DOE", "alice Johnson"]))
```
