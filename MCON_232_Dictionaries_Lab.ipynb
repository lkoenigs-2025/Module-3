```python
#LOOP THROUGH DICT:
#loop thru valus: use .values()
#or .items for key value pairs

print("Get keys")
for key in courses:
  print(key)

print()
print("Get keys and values")
for key in courses:
  print(key, ":", courses[key]) #gets the value once have keys

print()
print("Print values")
for value in courses.values():
  print(value)  #no way to get the keys in this way

#get value pairs using items()
print()
for key, value in courses.items():
  print(f"{key} : {value}")

#enumerate:
print()
for index, (key, value) in enumerate(courses.items()):
  print(f"{index}: {key} : {value}")  #useful for formatting
```

```python
student = {}
student[1] = "Leah"
student.get(1, "Leah")
print(student)

#if key doesnt exist then that key value pair will be added to dict
student.setdefault(2, "Reizi")
print(student)

#merges the two dict:
courses |= student
print(courses)

#we can pop
courses.popitem() #pops last item- no argument
print(courses)

courses.pop(3, "Ruchama") # if it doesnt exist nothing changes
courses.pop(1, "Leah") #pops off
print(courses)
```

```python
#will print the whole dict cuz references whole dict
LAS = courses
print(LAS)

#this will change mcon 232 from courses
LAS["MCON 232"] = "Change Course"
print(courses)

#to make a shallow copy:
print()
LAS = courses.copy()
LAS["MCON 232"] = "New Course"
print(courses)
print(LAS)
#This will only copy top level of dict: which can be a problem if nested dic/ nested list:
#So instead u wld do .deepcopy()
```

## Assignment: Majority Element (Easy)
Problem Given a list of integers nums, return the element that appears more   than n/2 times.
You may assume a majority element always exists.

Example
nums = [3, 2, 3] → 3
nums = [2, 2, 1, 1, 1, 2, 2] → 2

Goal Solve using a dictionary that maps: number -> count.


```python
def majority_elements(nums : list):
  counts = {}

#iterates thru list for each num
  for i in nums:
    #if does or doesnt exist, store as key in dict put 0 and add 1
    #if exists then u get the value and keep adding 1 so value for 4 changes
    counts[i] = counts.get(i, 0) + 1

    #now return element that is there more than 1
    if counts[i] > len(nums) // 2:
      return i

print(majority_elements([1,2,3,4,4,4,4,4]))
```
