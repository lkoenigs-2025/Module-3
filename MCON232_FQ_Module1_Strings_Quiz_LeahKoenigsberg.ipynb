## Question 1 — Kebab-case to Title (Est. 10 minutes)

Write a function:

```python
def kebab_to_title(s: str) -> str:
    ...
```

It should:
- take a string like `"a-little-princess"`
- return `"A Little Princess"`

Rules:
- The input may have extra spaces around it (strip them).
- Words are separated by `-`.
- Use string methods and loops (no fancy tricks).

Test:
- `"  anne-of-green-gables  "` → `"Anne Of Green Gables"`

````python
# Q1: Write your function here

def kebab_to_title(s: str) -> str:
    string = s.strip()
    string = string.replace("-"," ")
    string = string.title()
    return string


print(kebab_to_title("a-little-princess"))
print(kebab_to_title("  anne-of-green-gables  "))
````

## Question 2 — Count a Substring (Non-overlapping) (Est. 10 minutes)

Write a function:

```python
def count_substring(text: str, sub: str) -> int:
    ...
```

Return the number of times `sub` occurs in `text`, **without overlaps**.

Examples:
- `count_substring("aaaa", "aa")` → 2  (matches at 0–1 and 2–3)
- `count_substring("bananana", "ana")` → 2  (matches "ana" twice without overlaps)

````python
# Q2: Write your function here

def count_substring(text: str, sub: str) -> int:
  num = 0
  cnt = 0
  for i in range(len(text)):
    if text[cnt:cnt+len(sub)] == sub:
      num +=1
      cnt+= len(sub)
    else:
      num+=0
      cnt+=1
      if cnt > len(text):
        break
  return num



print(count_substring("aaaa", "aa"))
print(count_substring("bananana", "ana"))
````

## Question 3 — Phrase Palindrome (Est. 12–15 minutes)

Write a function:

```python
def is_phrase_palindrome(s: str) -> bool:
    ...
```

Return `True` if `s` is a palindrome **when you ignore**:
- spaces
- punctuation
- capitalization

Examples:
- `"Never odd or even"` → True
- `"Was it a rat I saw?"` → True
- `"Touro College"` → False

Hint: Build a cleaned string that keeps letters only.

````python
# Q3: Write your function here

def is_phrase_palindrome(s: str) -> bool:
    s = s.lower()
    s = s.replace(" ", "")
    new = s[::-1]

    punct = ",.?':;!{}[]()-_\""

    for p in punct:
      if p in s:
        s = s.replace(p,"")
        new = s[::-1]

    if new == s:
      return True

    else:
      return False


print(is_phrase_palindrome("Never odd or even"))
print(is_phrase_palindrome("Was it a rat I saw?"))
print(is_phrase_palindrome("Touro College"))
````

## Question 4 — CSV (Read without csv library) (Est. 15 minutes)

1) Run the next cell to write `students.csv`.
2) Read the file **without** using the `csv` library.
3) Return a list of lists, one entry per row
4) Compute, return and print the **average score** as a float rounded to 1 decimal.
5) make certain that code is separated by concern into appropriate functions.

CSV columns: `name,score,city`

Note: `score` must be converted to `int`.

````python
# PROVIDED: Run this cell first to write students.csv

def write_students_csv(filename: str = "students.csv") -> None:
    header = ["name", "score", "city"]
    rows = [
        ["Chana", 92, "New York"],
        ["Miriam", 85, "Brooklyn"],
        ["Leah", 88, "Queens"],
        ["Rivka", 95, "Lakewood"],
        ["Sara", 90, "Baltimore"],
    ]

    fout = open(filename, "w", encoding="utf-8", newline="")
    fout.write(",".join(header) + "\n")
    for row in rows:
        pieces = []
        for item in row:
            pieces.append(str(item))
        fout.write(",".join(pieces) + "\n")
    fout.close()

write_students_csv("students.csv")
print("Wrote students.csv")
````

````python
# Q4: Your work here (Starter Code)

def read_students_no_csv(filename: str = "students.csv") -> list[list[str]]:
    """
    Read the file without using the csv library.
    Return a list of lists, one entry per row
    """
    my_list = []
    students = []

    with open(filename, "r") as file:
      for line in file:
        line = file.readlines()

      for i in line:
        new = [i]
        new = i.strip()
        my_list.append(new)

      for i in range(len(my_list)):
        lst = my_list[i].split(",")
        students.append(lst)

    return students


def average_score(students: list[list[str]]) -> float:
    """
    Compute and return the average score.

    students is a list of lists:
        [name, score, city]
    Convert score to int before adding.
    """
    # TODO:
    # 1) loop through students
    # 2) convert row[1] to int
    # 3) compute total and count
    # 4) return total / count

    total = 0
    for i in range(len(students)):
      score = int(students[i][1])
      total+= score
    avg = total/len(students)
    return avg



students = read_students_no_csv("students.csv")
print(students)

avg = average_score(students)
print(f"Average score: {avg:.1f}")
````
